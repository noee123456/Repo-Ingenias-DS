12) ¿Cual es el promedio de expectativa de vida (Life expectancy) en general?

```python
# Obtengo el promedio de la columna
expvida['life_expectancy'].mean()
```

13) ¿Cual es el promedio de expectativa de vida (Life expectancy) por país?

```python
# Obtengo el promedio por pais
promedio_vida_pais = expvida.groupby('Country')[['life_expectancy']].mean()
promedio_vida_pais
```

14) ¿Cual es la expectativa de vida mas alta y mas baja? Usa la funcion de pandas y la funcion de NumPy y compara el resultado. ¿Que paises tienen esas expectativas de vida?

```python
# Maxima expectativa de vida usando numpy y otra con pandas
max_exp_vida_np = np.max(promedio_vida_pais)
print(max_exp_vida_np)
max_exp_vida_pd = promedio_vida_pais.max()
print(max_exp_vida_pd)
```

```python
# Minima expectativa de vida usando numpy y otra con panda
min_exp_vida_np = np.min(promedio_vida_pais)
print(min_exp_vida_np)
min_exp_vida_pd = promedio_vida_pais.min()
print(min_exp_vida_pd)
```

```python
# Muestro los resultados maximos con pandas y con numpy
print (f'Maxima expectativa de vida c/ Numpy: {max_exp_vida_np}')
print (f'Maxima expectativa de vida c/ Pandas: {max_exp_vida_pd}')
```

```python
# Muestro los resultados minimos
print (f'Minima expectativa de vida c/ Numpy: {min_exp_vida_np}')
print (f'Minima expectativa de vida c/ Pandas: {min_exp_vida_pd}')
```

```python
# Filtro las columnas para mostrar solo las filas que tengan valor
# igual al maximo
valores_maximos = promedio_vida_pais['life_expectancy']== max_exp_vida_np
promedio_vida_pais[valores_maximos]
```

```python
# Filtro las columnas para mostrar solo las filas que tengan valor
# igual al minimo
valores_minimos = promedio_vida_pais['life_expectancy']== min_exp_vida_np
promedio_vida_pais[valores_minimos]
```

18) ¿Cuantos paises presentaron algun caso de Sarampion (**Measles**) en el 2015? **_Pista_**: Lee acerca de la función [.count()](http://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.count.html?highlight=count#pandas.DataFrame.count)
